```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://imagery.tbportal.org/'
```

```python
r = requests.get(URL)
```

```python
r.status_code
```

```python
xml_doc = r.text
```

```python
soup = BeautifulSoup(xml_doc)
```

```python
print(soup.prettify())
```

```python
def save_file(url):
    file_path = os.path.dirname(url)
    if not os.path.exists(file_path):
        os.makedirs(file_path)
    
    
```

```python
for dcm in soup.find_all('key'):
    file_path = dcm.contents[0]
    dcm_url = os.path.join(URL,file_path)
    file_folder = os.path.dirname(file_path)
    if not os.path.exists(file_folder):
        os.makedirs(file_folder)
    
    r_dcm = requests.get(dcm_url, stream=True)
    
    with open(file_path, 'wb') as fp:
        for chunk in r_dcm.iter_content(chunk_size=512):
            if chunk:
                fp.write(chunk)
```
